```python
# ============================================
# 1. Import Libraries
# ============================================
import os, shutil, random
from pathlib import Path
import numpy as np
import cv2
import albumentations as A
from tqdm import tqdm
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
from pprint import pprint

# ============================================
# 2. Define Paths
# ============================================
BASE_PATH = Path("/kaggle/input/chest-xray-pneumonia/chest_xray")
TRAIN_PATH = BASE_PATH / "train"
VAL_PATH = BASE_PATH / "val"
TEST_PATH = BASE_PATH / "test"

NEW_TRAIN = Path("/kaggle/working/new_train")
NEW_VAL = Path("/kaggle/working/new_val")
BAL_TRAIN = Path("/kaggle/working/balanced_train")

# ============================================
# 3. Re-Split: Create New Train/Val
# ============================================
def prepare_split(src, new_train, new_val, test_size=0.15, seed=42):
    from sklearn.model_selection import train_test_split
    for cls in ["NORMAL", "PNEUMONIA"]:
        files = list((src/cls).glob("*.jpeg"))
        train_files, val_files = train_test_split(files, test_size=test_size, random_state=seed)
        for folder, group in [(new_train, train_files), (new_val, val_files)]:
            target = folder/cls
            target.mkdir(parents=True, exist_ok=True)
            for f in group:
                shutil.copy(f, target/f.name)

prepare_split(TRAIN_PATH, NEW_TRAIN, NEW_VAL)

# ============================================
# 4. Preprocessing Function
# ============================================
def preprocess_image(img_path, size=(224,224)):  # ResNet default size
    img = cv2.imread(str(img_path), cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, size)
    img = cv2.GaussianBlur(img, (3,3), 0)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    img = clahe.apply(img)
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)  # Convert to 3 channels
    img = img/255.0
    return img

# ============================================
# 5. Balance ONLY Training Set (Oversampling)
# ============================================
for cls in ["NORMAL", "PNEUMONIA"]:
    (BAL_TRAIN/cls).mkdir(parents=True, exist_ok=True)

counts = {cls: len(list((NEW_TRAIN/cls).glob("*.jpeg"))) for cls in ["NORMAL","PNEUMONIA"]}
max_count = max(counts.values())

augment = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.Rotate(limit=25, p=0.5),
    A.RandomBrightnessContrast(p=0.5),
    A.RandomGamma(p=0.3),
    A.ElasticTransform(alpha=1, sigma=50, alpha_affine=50, p=0.3),
    A.ShiftScaleRotate(shift_limit=0.1, scale_limit=0.1, rotate_limit=20, p=0.5)
])

for cls in ["NORMAL","PNEUMONIA"]:
    imgs = list((NEW_TRAIN/cls).glob("*.jpeg"))
    for f in imgs:
        img = preprocess_image(f)
        save = (BAL_TRAIN/cls)/f.name
        cv2.imwrite(str(save), (img*255).astype(np.uint8))
    cur = len(imgs)
    while cur < max_count:
        img_path = random.choice(imgs)
        img = cv2.imread(str(img_path))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img, (224,224))
        img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
        aug_img = augment(image=img)['image']
        save = (BAL_TRAIN/cls)/f"aug_{cur}.jpeg"
        cv2.imwrite(str(save), aug_img)
        cur += 1

# ============================================
# 6. Create TensorFlow Datasets
# ============================================
IMG_SIZE = (224,224)
BATCH = 32

def make_dataset(path):
    return tf.keras.preprocessing.image_dataset_from_directory(
        path, image_size=IMG_SIZE, batch_size=BATCH, label_mode='binary', shuffle=True
    )

train_ds = make_dataset(BAL_TRAIN)
val_ds = make_dataset(NEW_VAL)
test_ds = make_dataset(TEST_PATH)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(AUTOTUNE)
val_ds = val_ds.prefetch(AUTOTUNE)
test_ds = test_ds.prefetch(AUTOTUNE)

# ============================================
# 6b. Compute Class Weights
# ============================================
y_train = np.concatenate([y.numpy().flatten() for x, y in train_ds], axis=0)
class_weights_values = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(enumerate(class_weights_values))
print("Class weights:", class_weight_dict)

# ============================================
# 7. Build Model (ResNet101 Fine-Tuned)
# ============================================
from tensorflow.keras.applications import ResNet101
from tensorflow.keras.applications.resnet import preprocess_input

base_model = ResNet101(weights='imagenet', include_top=False, input_shape=(224,224,3))
base_model.trainable = True
for layer in base_model.layers[:-50]:  # freeze early layers
    layer.trainable = False

inputs = tf.keras.Input(shape=(224,224,3))
x = preprocess_input(inputs)
x = base_model(x, training=True)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(512, activation='relu')(x)
x = tf.keras.layers.Dropout(0.5)(x)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.3)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs, outputs)
model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# ============================================
# 8. Train Model with Callbacks
# ============================================
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)
lr_sched = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)

history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
    class_weight=class_weight_dict,
    callbacks=[early_stop, lr_sched]
)

# ============================================
# 9. Plot Training History
# ============================================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Val Accuracy', marker='o')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss', marker='o')
plt.plot(history.history['val_loss'], label='Val Loss', marker='o')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================
# 10. Evaluate Dataset Function
# ============================================
def evaluate_dataset(model, dataset, dataset_name="Dataset"):
    y_true, y_probs = [], []
    for imgs, labels in dataset:
        probs = model.predict(imgs, verbose=0).flatten()
        y_probs.extend(probs)
        y_true.extend(labels.numpy().flatten())
    
    y_true = np.array(y_true)
    y_pred = (np.array(y_probs) > 0.5).astype(int)
    
    acc = accuracy_score(y_true, y_pred)
    report = classification_report(y_true, y_pred, target_names=["NORMAL","PNEUMONIA"], output_dict=True)
    
    print(f"\n📌 {dataset_name} Results:")
    print(f"Accuracy: {acc:.4f}")
    print(f"Weighted Precision: {report['weighted avg']['precision']:.4f}")
    print(f"Weighted Recall:    {report['weighted avg']['recall']:.4f}")
    print(f"Weighted F1-score:  {report['weighted avg']['f1-score']:.4f}")
    
    pprint(report)
    
    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=["NORMAL","PNEUMONIA"], yticklabels=["NORMAL","PNEUMONIA"])
    plt.title(f"{dataset_name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_true, y_probs)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
    plt.plot([0,1],[0,1],'--',color='gray')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"{dataset_name} ROC Curve")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    return y_true, y_pred, y_probs, report

# ============================================
# 11. Evaluate Train, Validation, and Test Sets
# ============================================
evaluate_dataset(model, train_ds, "Train Set")
evaluate_dataset(model, val_ds, "Validation Set")
evaluate_dataset(model, test_ds, "Test Set")
```
